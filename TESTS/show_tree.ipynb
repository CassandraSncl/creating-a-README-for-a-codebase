```python
import os
```

```python
def show_tree(root):
    for dirpath, dirnames, filenames in os.walk(root):
        level = dirpath.replace(root, "").count(os.sep)
        indent = " " * 4 * level
        print(f"{indent}{os.path.basename(dirpath)}/")
        subindent = " " * 4 * (level + 1)
        for f in filenames:
            print(f"{subindent}{f}")

show_tree(".") # <-- Remplacer "." par le chemin du répertoire souhaité
```

```python
# Extensions à classer
code_ext = {".py", ".java", ".cpp", ".c", ".js", ".html", ".css", ".ipynb", ".ts"}
doc_ext  = {".txt", ".pdf", ".doc", ".docx", ".xls", ".xlsx", ".ppt", ".pptx", ".md",".csv"}

def classify_files(root):
    codes, docs, others = [], [], []
    
    for dirpath, _, filenames in os.walk(root):
        for f in filenames:
            ext = os.path.splitext(f)[1].lower()
            full_path = os.path.join(dirpath, f)
            if ext in code_ext:
                codes.append(full_path)
            elif ext in doc_ext:
                docs.append(full_path)
            else:
                others.append(full_path)
    
    # Affichage de l'arborescence
    print("\n=== Arborescence ===")
    show_tree(root)
    
    # Résultats classés
    print("\n=== Codes ===")
    print("\n".join(codes) if codes else "Aucun fichier code trouvé")

    print("\n=== Documents ===")
    print("\n".join(docs) if docs else "Aucun document trouvé")

    print("\n=== Autres ===")
    print("\n".join(others) if others else "Aucun fichier autre trouvé")

# Exemple d'appel
classify_files("D:\\33611\\Documents\\MASTER\\M2\\Données Séquentielles\\projet")  # "." = dossier courant
```
